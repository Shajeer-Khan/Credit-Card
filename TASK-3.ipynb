```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('C:/Users/Shajeer Waqar/Downloads/archive/creditcard.csv')  
```

```python
# Display basic information about the dataset
print(data.info())
```

```python
# Summary statistics of numerical columns
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Visualizing class distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=data)
plt.title('Class Distribution')
plt.show()
```

```python
# Checking correlation among features
plt.figure(figsize=(12, 10))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualizing distribution of some features for both classes
features_to_plot = ['V1', 'V2', 'V3', 'V4', 'V10', 'V14', 'V17', 'Amount']

for feature in features_to_plot:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=data, x=feature, hue='Class', kde=True, bins=50, palette='viridis')
    plt.title(f'Distribution of {feature} by Class')
    plt.show()
```
